## Formatting the Columns Names

```python
# a general function to convert any camelCase to snake_case
def snakecase_formatter(word):
    modded_word = ""
    for i,v in enumerate(word):
        if i != 0 and i != "_" and v.isupper():
            modded_word = modded_word + "_" + v.lower()
        else:
            modded_word = modded_word + v.lower()
    return modded_word
```

We will reformat the columns names, we will change yearOfRegistration, monthOfRegistration, notRepairedDamage, and dateCreated columns' names to be easier to understand. Alongside of that, we will also reformat the columns names with our function

```python
columns_names = list(autos.columns)

d = {
    "yearOfRegistration": "registration_year",
    "monthOfRegistration": "registration_month",
    "notRepairedDamage": "unrepaired_damage",
    "dateCreated": "ad_created"
    }

for i,v in enumerate(columns_names):
    if v in d.keys():
        columns_names[i] = d[v]
    else:
        columns_names[i] = snakecase_formatter(v)

autos.columns = columns_names
```

## Converting German to English

```python
from googletrans import Translator
```

```python
# creating a translated word dictionary
german_label = ["seller", "offer_type", "abtest", 
                "vehicle_type", "gearbox",
                "unrepaired_damage", "fuel_type"]
translated_unique = {}
for label in german_label:
    for word in (autos[label].unique()):
        if type(word) != float:
            translator = Translator()
            # telling the translator to translate from german
            translated_unique[word] = (translator.translate(word, src = "de").text)
    # mapping the translation
    autos[label] = autos[label].map(translated_unique)
    translated_unique = {}
autos
```

Because eBay bidding starts at one dollar, we can remove listings with prices of zero. Even if you didn't know anything about eBay, you can come to the conclusion that no normal person will list their car at price zero. Looking at the upperend of the prices, we can see that the outliers are after 350,000. We say this because there is a crazy hike in price after 350,000 whereas the price steadily increases before. The reason why we are not using any outlier algorithm like z-score and interquartile range is because eBay is an auction site. If we use any of those algorithm, the super low prices and high prices will be cut out from the dataset resulting in loss of potentially valulable insights. For example, lower end cars and cars with high mileage may tend to have low starting bidding prices while higher end cars will have higher starting bidding price. ## Deeper Dive into Price vs Mileage

```python
autos["odometer_km"].value_counts()
```

```python
# find avg price for each mileage in the listing
unique_mileage = autos["odometer_km"].unique()
mileage_price = {}

for mileage in unique_mileage:
    avg_price = np.mean(autos[autos["odometer_km"] == mileage]["price"])
    mileage_price[mileage] = avg_price
mileage_price
```

```python
x, y = zip(*sorted(mileage_price.items()))
plt.scatter(x, y)
```

As we can see from the avg prices, there is an inverse relationship between mileage and price. As the mileage increases, the price decreases and vice versa. Cars with higher mileages tend to require more maintenance than cars with low mileages. If the prices were to not be affected by the mileages, buyers would always go for cars with lower mileages. It doesn't make sense to buy a car with the higher mileage that would require more money and time to maintain. Therefore, the price is lowered to offset the increase in opportunity cost of buying a high mileage car. However, we can see that there is an anomally in the avg price of cars with mileages of 5000. If the inverse relationship was to hold, cars with a mileage of 5000 should have the highest selling price.

## Conclusion

```python
for m in np.sort(autos["odometer_km"].unique()):
    std_registration_year = autos[autos["odometer_km"] == m]["registration_year"].describe()["std"]
    std_price = autos[autos["odometer_km"] == m]["price"].describe()["std"]
    mean_price = autos[autos["odometer_km"] == m]["price"].describe()["mean"]
    
    print(f"std of Registration Year({m} mileage): {std_registration_year}")
    print(f"std of Price({m} mileage): {std_price}")
    print(f"mean of Price({m} mileage): {mean_price}\n")
```

We learned that if we hold the brand constant, the registration year directly affects the price of a car. What if the brands were not constant? Because the registration year dictates a car type(ex: Audi A4 2010 and Audi A4 2013 are very different in terms of price, comestic, fuel efficieny, and etc), the closer the registration years of the car the closer the prices are to each other. The cars don't have to be a same brand because a 1999 car will always be cheaper than 2017 car. As long as the std of the registration years of cars in each mileages are not too high(in this case, std <= 10), the inverse relationship of mileage and price holds true. In addition, each listings must contain about similar amount of each brand for the relationship to hold. A listing of ferraris will always have a higher avg price than a listing with Toyotas regardless of mileage.
